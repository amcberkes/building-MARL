```python
import pickle
import pprint

# Load the dataset
with open('hvac_dataset.pkl', 'rb') as f:
    hvac_dataset = pickle.load(f)

# Check the number of entries in the dataset
print(f"Total number of trajectories in the dataset: {len(hvac_dataset)}\n")

# Peek at the first few entries
num_entries_to_show = 3  # Number of entries to display
print(f"Displaying the first {num_entries_to_show} trajectories:\n")

for i, entry in enumerate(hvac_dataset[:num_entries_to_show]):
    print(f"Trajectory {i+1}:")
    pprint.pprint(entry)
    print("\n" + "="*80 + "\n")
```
